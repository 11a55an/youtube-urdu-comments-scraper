```python
import pandas as pd
import re
from googleapiclient.discovery import build

# Set up the YouTube API
api_key = 'API KEY'
youtube = build('youtube', 'v3', developerKey=api_key)

def is_urdu(comment):
    # Regular expression pattern to identify Urdu script characters
    urdu_pattern = re.compile(r'[\u0600-\u06FF]+')
    return bool(urdu_pattern.search(comment))

def get_video_comments(video_id):
    comments = []
    nextPageToken = None

    while True:
        response = youtube.commentThreads().list(
            part='snippet',
            videoId=video_id,
            textFormat='plainText',
            pageToken=nextPageToken
        ).execute()

        for item in response['items']:
            topLevelComment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            if is_urdu(topLevelComment):
                comments.append(topLevelComment)

            # Check if there are replies for the current comment
            if 'replies' in item:
                for reply in item['replies']['comments']:
                    replyComment = reply['snippet']['textDisplay']
                    if is_urdu(replyComment):
                        comments.append(replyComment)

        nextPageToken = response.get('nextPageToken')
        if not nextPageToken:
            break

    return comments

def get_row():
    df2 = pd.read_excel("comments.xlsx", sheet_name="Sheet1")
    num_rows = len(df2)
    return num_rows

def save_comments_to_spreadsheet(video_id, file_path):
    comments = get_video_comments(video_id)
    df = pd.DataFrame({'Comments': comments})
    startRow = get_row()
    print (startRow)
    with pd.ExcelWriter('comments.xlsx', mode="a", if_sheet_exists="overlay") as writer:
        df.to_excel(writer, startrow = startRow+1, header=False, index=False)
    print(len(df))

if __name__ == "__main__":
    video_id = 'PD-zh2omR_A'
    file_path = 'comments.xlsx'
    save_comments_to_spreadsheet(video_id, file_path)
```
